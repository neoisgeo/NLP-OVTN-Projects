# 2. Get Sentiment , Emotion, Subjectivity Score

### Sentiment Score 

```python
def get_sentiment(texts, idx):
    # Check if the Language is English or not

    try:
        language = detect(texts)
    except:
        return
        
    if language != 'en':
        return
    
    #Check if the word length is more than 5
    
    wordLength = len(texts.split()) 
    
    if int(wordLength) < 5:
        return
    
    response = natural_language_understanding.analyze (
        text = texts,
        features = Features(sentiment=SentimentOptions()
                           )).get_result()
    
    jsonResponse = json.dumps(response, indent=2)
    
    X = json.loads(jsonResponse)
       
    sentiment = X['sentiment']['document']['score']
    SentimentScore.append(sentiment)
```

```python
SentimentScore = []

for idx in range(len(df_data_4)):
    get_sentiment(df_data_4["clean_content"].iloc[idx], idx)
```

```python
df_data_4['sentiment_document'] = SentimentScore
df_data_4.head()
```

### Emotion Score

```python
def emotionScore(texts):
    # Check if the Language is English or not

    try:
        language = detect(texts)
    except:
        return
        
    if language != 'en':
        return
    
           # try:
        #  subtypes.append(i['disambiguation']['subtype'])
       # except:
         # subtypes.append("")
    
    #Check if the word length is more than 5
    
    wordLength = len(texts.split()) 
    
    if int(wordLength) < 5:
        return
        
    # Feed the text to NLU and get the Sentiment and Emotions
    
    response = natural_language_understanding.analyze(
        text=texts,
        features=Features(emotion=EmotionOptions()),
        language="en").get_result()
    
    jsonResponse = json.dumps(response, indent=2)
    
    X = json.loads(jsonResponse)
    
    sadness.append(X["emotion"]["document"]["emotion"]["sadness"])
    joy.append(X["emotion"]["document"]["emotion"]["joy"])
    fear.append(X["emotion"]["document"]["emotion"]["fear"])
    disgust.append(X["emotion"]["document"]["emotion"]["disgust"])
    anger.append(X["emotion"]["document"]["emotion"]["anger"])
```

```python
sadness = []
joy = []
fear = []
disgust = []
anger = []


for idx in range(len(df_data_4)):
    emotionScore(df_data_4["clean_content"].iloc[idx])

df_data_4["sadness"] = sadness
df_data_4["joy"] = joy
df_data_4["disgust"] = disgust
df_data_4["fear"] = fear
df_data_4['anger'] = anger

df_data_4.head()
```
